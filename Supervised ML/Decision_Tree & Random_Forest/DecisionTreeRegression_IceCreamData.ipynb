Decision Tree Regression on IceCreamData

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('./Data/IceCreamData.csv')
x=data['Temperature'].values
y=data['Revenue'].values
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.05)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor().fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
y_pred=regressor.predict(x_test.reshape(-1,1))
```

```python
df = pd.DataFrame({'Real Values':y_test.reshape(-1),'Predicted Values':y_pred.reshape(-1)})
df.head()
```

```python
x_grid=np.arange(min(x),max(x),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x_test,y_test,c='red',label='Real')
plt.scatter(x_test,y_pred,c='green',label='predicted')
plt.title('Decision Tree Regression')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [15, 9]
plt.plot(x_grid,regressor.predict(x_grid));
plt.title('Decision Tree Regression')
plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.show()
```
