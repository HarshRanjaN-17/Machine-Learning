Linear Regression Using Gradient Descent Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('./Data/dataSet.csv')
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.rcParams['figure.figsize']=(12.0,9.0)
plt.scatter(x,y,s=10)
plt.show()
```

```python
#   Building the model
m=c=0
l=0.0001    #learning rate
epochs=1000 #no. of iterations to perform gradient descent
n=float(len(x)) #no. of element in x

for i in range(epochs):
    y_pred=m*x+c
    d_m=(-2/n)*sum(x*(y-y_pred))    #derivative wrt m
    d_c=(-2/n)*sum(y-y_pred)        #derivative wrt c

    m-=l*d_m
    c-=l*d_c
print(f'slope : {m}\nIntercept : {c}')
```

```python
#   Making Predictions
y_pred=m*x+c
plt.scatter(x,y)
plt.plot(x,y_pred,c='orange')
#plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],c='orange')
plt.show()
```
