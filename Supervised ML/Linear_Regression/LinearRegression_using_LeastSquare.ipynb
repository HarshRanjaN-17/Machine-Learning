Linear Regression Using Least Square

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)
```

```python
data=pd.read_csv('./Data/dataSet.csv')
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(x,y)
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)

num=den=0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2
m=num/den
c=y_mean-m*x_mean
print(f'm = {m}\tc = {c}'); # y=mx+c
```

```python
y_pred=m*x+c
plt.scatter(x,y)
plt.plot(x,y_pred,'orange')
```
