Linear Regression implemention using scipy module

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./Data/SimpleLinearRegression.csv')
X=df['SAT']
Y=df['GPA']
# print(X.head())
# print(Y.head())
```

```python
slope,intercept,r,p,std_err=stats.linregress(X,Y)
print("Slope : {:.6f}".format(slope))
print(f'Intercept : {intercept:.6f}')
print('Correlation Coefficient : ',r)
print('p_value : ',p)
print('Standard Error : ',std_err)
```

```python
def myfun(x):
    return slope*x+intercept

myModel=list(map(myfun,X))

plt.scatter(X,Y)
plt.title('Linear Regression')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.plot(X,myModel,c='orange')
plt.grid(True)
```
