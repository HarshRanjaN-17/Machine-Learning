```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
print('Image data shape : ',digits.data.shape)
print('Label data shape : ',digits.target.shape)
```

```python
plt.figure(figsize=(20,5))
for idx, (image,label) in enumerate(zip(digits.data[0:10], digits.target[0:10])):
    plt.subplot(1,10,idx+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title('Training : %i\n'%label,fontsize=10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(digits.data , digits.target , test_size = 0.2 , random_state = 2)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression().fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
score=regressor.score(x_test,y_test)
score
```

```python
test_list=[y_pred,y_test]
# for x,y in zip(*test_list):
#     print(x,y)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
```
